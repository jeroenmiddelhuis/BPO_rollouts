```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

nr_rollouts = [50, 100, 200]
nr_steps_per_rollout = [25, 50, 100]
tau_multiplier = [0.1, 0.25, 0.5, 1.0]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

parameter_performance = []

for nor in nr_rollouts:
    for nsp in nr_steps_per_rollout:
        for tm in tau_multiplier:
            if nor == 200 and nsp == 25 and tm == 1.0: # We don't have this file
                continue
            df = pd.read_csv(f"./results_tuning/pi_slow_server_{nor}_{nsp}_{tm}.txt")
            mean = df["mean"][0]
            parameter_performance.append((nor, nsp, tm, mean))
            ax.scatter(nor, nsp, tm, c=mean, cmap='viridis')

ax.set_xlabel('Number of Rollouts')
ax.set_ylabel('Number of Steps per Rollout')
ax.set_zlabel('Tau Multiplier')
plt.show()
print('Number of Rollouts, Number of Steps per Rollout, Tau Multiplier, Mean')
parameter_performance.sort(key=lambda x: x[3], reverse=True)
for i in parameter_performance:
    print(i)
```
